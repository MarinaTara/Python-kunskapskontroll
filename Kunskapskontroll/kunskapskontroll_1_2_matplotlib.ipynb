```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

1. Object-Oriented approach
2. pyplot/MATLAB style
In general it is recommended to use the Object-Oriented approach, especially for more complicated plots. 
The pyplot/MATLAB approach can be used for quick interactive work.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: a top-level container that holds all the elements of a plot or visualization. The whole figure.
Axes: represents the entire plotting area on a figure, and usually includes 2 (or 3 in the case of 3D) Axis objects.
Axis: sets the scale and limits and generate ticks (the marks on the Axis) and ticklabels (strings labeling the ticks).
Artist: basically everything we see on the Figure is an Artist(Axes, Axis, Text objects etc.)

### When plotting in Matplotlib, what is the expected input data type?

Plotting functions expect numpy.array or numpy.ma.masked_array as input, or objects that can be passed to numpy.asarray.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Object-Oriented approach
x = np.linspace(-4, 4, 100)
y = x**2
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('y = x^2', fontsize=14, color='red')
ax.set_xlabel('x', fontsize=13)
ax.set_ylabel('y', fontsize=13)
```

```python
# pyplot/MATLAB approach
plt.plot(x, y)
plt.title("y = x^2", fontsize=14, color='blue')
plt.xlabel('x', fontsize=13)
plt.ylabel('y', fontsize=13)
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Creating 2 subplots
fig, axs = plt.subplots(1, 2, layout='constrained')

axs[0].scatter(random_data_x, random_data_y)
axs[0].set_title('Scatter Plot', fontsize=14, color='purple')

axs[1].bar(names, values)
axs[1].set_title('Bar Plot', fontsize=14, color='turquoise')
```
